```python
import numpy as np

def my_logistic_eq(t, P, r, K):
    
    dP_dt = r * P * (1 - P / K)
    return dP_dt
```

```python
from scipy.integrate import solve_ivp

P0 = 100
r = 0.1
K = 1000

t_span = (0, 50)

sol = solve_ivp(my_logistic_eq, t_span, [P0], args = (r, K), t_eval = np.linspace(0, 50, 100))

import matplotlib.pyplot as plt
plt.plot(sol.t, sol.y[0], label ='Population')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Logistic Equation Solution')
plt.legend()
plt.grid(True)
plt.show
```
