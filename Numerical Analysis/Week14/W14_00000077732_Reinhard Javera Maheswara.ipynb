```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk membuat gelombang sinus
def sinus_wave(freq, amplitude, phase_shift, duration, sampling_rate):
    t = np.arange(0, duration, 1/sampling_rate)
    return amplitude * np.sin(2 * np.pi * freq * t + phase_shift)

# Parameter sinyal 1
freq_1 = 5 # Hz
amplitude_1 = 3
phase_shift_1 = 3

# Parameter sinyal 2
freq_2 = 2 # Hz
amplitude_2 = 2
phase_shift_2 = -2

# Laju sampling
sampling_rate = 100 # Hz

# Durasi sinyal
duration = 2 # detik

# Hasilkan sinyal 1 dan sinyal 2
signal_1 = sinus_wave(freq_1, amplitude_1, phase_shift_1, duration, sampling_rate)
signal_2 = sinus_wave(freq_2, amplitude_2, phase_shift_2, duration, sampling_rate)

# Tambahkan sinyal 1 dan sinyal 2 bersama-sama
combined_signal = signal_1 + signal_2

# Plot sinyal
plt.figure(figsize=(10, 6))
plt.plot(np.arange(0, duration, 1/sampling_rate), combined_signal)
plt.title('Gabungan Sinyal Sinus')
plt.xlabel('Waktu (s)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk mengeksplorasi perbedaan laju sampling
def explore_sampling_rate(freq_1, amplitude_1, phase_shift_1, freq_2, amplitude_2, phase_shift_2, duration, sampling_rates):
    for sr in sampling_rates:
        signal_1 = sinus_wave(freq_1, amplitude_1, phase_shift_1, duration, sr)
        signal_2 = sinus_wave(freq_2, amplitude_2, phase_shift_2, duration, sr)
        combined_signal = signal_1 + signal_2
        plt.figure(figsize=(10, 3))
        plt.plot(np.arange(0, duration, 1/sr), combined_signal)
        plt.title(f'Gabungan Sinyal Sinus (Sampling Rate = {sr} Hz)')
        plt.xlabel('Waktu (s)')
        plt.ylabel('Amplitudo')
        plt.grid(True)
        plt.show()
        
# Laju sampling yang ingin dieksplorasi
sampling_rates = [5, 10, 20, 50, 100]
# Panggil fungsi untuk mengeksplorasi perbedaan laju sampling
explore_sampling_rate(freq_1, amplitude_1, phase_shift_1, freq_2, amplitude_2, phase_shift_2, duration, sampling_rates)
```
