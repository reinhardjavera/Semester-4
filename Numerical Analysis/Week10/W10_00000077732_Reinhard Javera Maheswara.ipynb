<h3>Code Forward Difference</h3>

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.cos(x)

def f_prime(x):
    return -np.sin(x)

def forward_difference(f, x, h):
    return (f(x + h) - f(x)) / h

x_values = np.linspace(0, 2*np.pi, 100)

true_derivative = f_prime(x_values)

h = 0.1

numerical_derivative = forward_difference(f, x_values, h)

error = np.abs(true_derivative - numerical_derivative)

max_error = np.max(error)

plt.figure(figsize=(10, 6))

plt.plot(x_values, true_derivative, label="Analytical Derivative", color='blue')
plt.plot(x_values, numerical_derivative, label="Numerical Derivative (Forward Difference)", linestyle='--', color='orange')
plt.scatter(x_values, numerical_derivative, color='orange', s=10)
plt.title("Numerical Differentiation of f(x) = cos(x)")
plt.xlabel("x")
plt.ylabel("Derivative")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(max_error)
```

<h3>Code Backward Difference</h3>

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.cos(x)

def f_prime(x):
    return -np.sin(x)

def backward_difference(f, x, h):
    return (f(x) - f(x - h)) / h

x_values = np.linspace(0, 2*np.pi, 100)

true_derivative = f_prime(x_values)

h = 0.1

numerical_derivative = backward_difference(f, x_values, h)

error = np.abs(true_derivative - numerical_derivative)

max_error = np.max(error)

plt.figure(figsize=(10, 6))

plt.plot(x_values, true_derivative, label="Analytical Derivative", color='blue')
plt.plot(x_values, numerical_derivative, label="Numerical Derivative (Backward Difference)", linestyle='--', color='orange')
plt.scatter(x_values, numerical_derivative, color='orange', s=10)
plt.title("Numerical Differentiation of f(x) = cos(x)")
plt.xlabel("x")
plt.ylabel("Derivative")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(max_error)
```
